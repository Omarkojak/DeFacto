```python
import json

## line 1584 problem institution file

def read_file(filepath = "./relation-extraction-corpus/20130403-institution.json"):
	with open(filepath) as fp:
		line = fp.readline()
		cnt = 1
		while line:
			if cnt == 1584 or cnt == 4803 or cnt == 6921 or cnt == 13051:
				line = fp.readline()
				cnt += 1
				continue

			text = json.loads(line.strip())

			if len(text["evidences"]) > 1:
				print(cnt)

			line = fp.readline()
			cnt += 1

read_file()
```
